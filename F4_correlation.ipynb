What I want to do now is take the Adjusted Close column in all the csv files and form one big DataFrame with them, where the 'Date' is the the index (rows) and the different companies are the columns.

```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')

```

```python
def combine_data():
    with open("sp500tickers.pickle", "rb") as f:
        tickers = pickle.load(f)
    main_df = pd.DataFrame()
    
    for count, ticker in enumerate(tickers):
        df = pd.read_csv('stock_dfs/{}.csv'.format(ticker))
        df.set_index('Date', inplace=True)
        df.rename(columns={'Adj Close':ticker}, inplace=True)
        df.drop(['Open','High','Low','Close','Volume'], axis=1, inplace=True) #axis default is 0, but this returns an error
        
        if main_df.empty: #now we join all these 'Adj Close' DataFrames together.
             main_df = df
        else:
            main_df = main_df.join(df, how='outer') #outer joining tables like in sql (so we don't lose data if some rows are missing from some for some tickers).
        if count%10 == 0:
            print(count) # So every 10th ticker in the list gets printed just to check we get roughly all the companies (without having to go through the entire list)
    print(main_df.head())
    main_df.to_csv('sp500_joined_closes.csv')
    
```

```python
combine_data()
```
